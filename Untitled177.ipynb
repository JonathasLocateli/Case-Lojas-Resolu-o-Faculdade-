```python
aluno_Jonathas = "Jonathas Locateli"
print(f"Seja bem vindo a loja do {aluno_Jonathas} \n") #IDENTIFICADOR PESSOAL
valor_unitario = float(input('Entre com o valor do produto: '))  # Define a variavel valor_unitario
# com o valor de cada produto
quantidade = int(input('Entre com o valor da quantidade: '))  # Define a variavel quantidade
# com a quantidade de produtos
valor_frete = 0  # Variavel que receberá o valor do frete dependendo da quantidade de
# produtos inseridos
if 0 <= quantidade < 11:
   valor_frete = 30
elif 11 <= quantidade < 101:
   valor_frete = 60
elif 101 <= quantidade < 1001:
   valor_frete = 120
else:
   valor_frete = 240
valor_parcial = float(valor_unitario * quantidade)  # Define a variavel valor_parcial com o valor
# parcial da multiplicação entre o valor_unitario * a quantidade
valor_total = float(valor_parcial + valor_frete)  # Define a variavel valor_total com o valor
# total da soma entre o valor_parcial + o valor_frete
print('O valor sem frete foi: R$ {:.2f}'.format(valor_parcial))  # imprime o valor_parcial formatado
# com duas casas decimais após o ponto
print('O valor com desconto foi: R$ {:.2f} '.format(valor_total) + '(frete de R$ {:.2f})'.format(valor_frete))
# imprime o valor_total formatado com duas casas decimais após o ponto
```

```python
aluno_Jonathas = "Jonathas Locateli"
print(f"Seja bem vindo a sorveteria do {aluno_Jonathas} \n") #IDENTIFICADOR PESSOAL
# tabela
tamanho = ["P", "M", "G"]
codigos = {"TR": [6.00, 10.00, 18.00],
          "ES": [7.00, 12.00, 21.00],
          "PR": [8.00, 14.00, 21.00]}
compra = []
# pedindo os dados
while True:
   qual_tamanho = input("Qual o tamanho do sorvete desejado? ")
   qual_sabor = input("Qual o código do sorvete desejado? ")
   if qual_tamanho in tamanho and qual_sabor in codigos:
       pedido = codigos[qual_sabor][tamanho.index(qual_tamanho)]
       compra.append(pedido)
       algo_mais = input ("Deseja pedir algo mais?"
                          "\nDigite S para sim ou N para não. ")
       if algo_mais == "S":
           continue
       else:
           break
   else:
       print("TAMANHO ou CÓDIGO inválido(s)")
       continue
print ("Valor total da compra:", "R$",sum(compra))
```

```python
aluno_Jonathas = "Jonathas Locateli"
# IDENTIFICADOR
print(f"Seja bem vindo ao Programa Serviços de Limpeza do {aluno_Jonathas}")
tipos_de_lavagem = {
    "b": ["Básica - Indicada para sujeiras semanais ou quinzenais", 1],
    "c": ["Completa (30% a mais) - Indicada para sujeiras antigas e/ou não rotineiras", 1.30]
}
lista_adicionais = { 
    0: ["Não desejo mais nada (encerrar)",              0],
    1: ["Passar 10 peças de roupas - R$ 10,00",        10],
    2: ["Limpeza de 1 Forno/Micro-ondas - R$ 12,00",   12],
    3: ["Limpeza de 1 Geladeira/Freezer - R$ 20,00",   20]
}

def metragem_limpeza(): # essa função pede para o usuario que entre com o numero da metragem da cas

    valor = 0
    while True:
        try:
            metragem = float(input("Entre com a metragem da casa (m): ").strip().upper())
            
            if metragem >= 30 and metragem < 300:
                valor = 60 + 0.3 * metragem
            elif metragem >= 300 and metragem < 700:
                valor = 120 + 0.5 * metragem
            else:
                print("\nNão aceitamos casas com metragem menor que 30m² ou maior que 700m²!\n")

            if valor > 0:
                break

        except ValueError:
            print("\nDigite um valor de metragem válido.\n")

    return valor #retorna o valor calculado

def tipo_limpeza(): # essa função solicita ao usuário o tipo de limpeza, retornando o valor multiplicador associado
    multiplicador = None
    while True:
        try:
            print("Entre com o tipo de limpeza:")
            for key, values in tipos_de_lavagem.items():
                print(f"{key} - {values[0]}")

            opcao = input().strip().lower()
            
            if opcao not in tipos_de_lavagem.keys():
                print("\nOpção inválida\n")
            else:
                multiplicador = tipos_de_lavagem[opcao][1]
                print(f"Você selecionou a opção {tipos_de_lavagem[opcao][0]}")
                break                
        except ValueError:
            print("\nDigite um opcao válida.\n")

    return multiplicador

def adicional_limpeza():
  #   Solicita ao usuário entrar com o número do adicional desejado, cada adicional é somado e retorna o valor da soma dos adicionais selecionados

    total_adicionais = 0
    while True:
        try:
            print("Deseja mais algum adicional?")
            for key, values in lista_adicionais.items():
                print(f"{key} - {values[0]}")

            opcao = int(input())
            
            if opcao not in lista_adicionais.keys():
                print("\nOpção inválida\n")
            elif opcao == 0:
                break                
            else:
                total_adicionais += lista_adicionais[opcao][1]

        except ValueError:
            print("\nDigite um adicional válido.\n")

    return total_adicionais

print("*"*80)
print(f"{' Menu 1 de 3 - Metragem Limpeza ':-^80}")
valor = metragem_limpeza()
print("*"*80)
print(f"{' Menu 2 de 3 - Tipo de Limpeza ':-^80}")
tipo = tipo_limpeza()
print("*"*80)
print(f"{' Menu 3 de 3 - Adicionais de Limpeza ':-^80}")
adicionais = adicional_limpeza()
print("*"*80)
total = (valor * tipo) + adicionais

print(f"TOTAL (R$): {total:.2f} (metragem = {valor:.2f} * tipo = {tipo:.2f} + adicionais = {adicionais:.2f})")
print("*"*80)
```

```python
aluno_Jonathas = "Jonathas Locateli"
# IDENTIFICADOR
print(f"Seja bem vindo ao Controle de Funcionários do {aluno_Jonathas}") # identificador pessoal
menu_principal = {
    1: "Cadastrar Funcionário",
    2: [
        "Consultar Funcionário",
        {
            1: "Consultar Todos os Funcionários",
            2: "Consultar Funcionário por Id",
            3: "Consultar Funcionário(s) por Setor",
            4: "Retornar",
        },
    ],
    3: "Remover Funcionário",
    4: "Sair",
}
def imprime_menu(menu: dict):
    """
    Imprime o menu
    """

    for key, val in menu.items():
        print(f"{key} - {val if type(val) == str else val[0]}")

    print("-" * 30)


def mostrar_menu(menu: dict) -> int:
    """
    Exibe o menu, solicita ao usuário uma opção e retorna o número da opção
    """

    print("*" * 80)
    print(f"{' MENU PRINCIPAL ':-^80}")
    imprime_menu(menu)

    while True:
        try:
            opcao = int(input("Escolha a opção desejada: "))

            if opcao in menu:
                break
            else:
                print("\nOpção inválida!\n")

        except ValueError:
            print("\nAtenção! Você digitou uma opção inválida.\n")

    return opcao


def cadastrar_funcionario(id: int):
    """
    Cadastra um funcionário recebendo um inteiro como argumento, solicita ao usuário um nome, 
    um departamento e um salário e, em seguida, adiciona as informações a um dicionário.
    """

    print("*" * 80)
    print(f"{' MENU CADASTRAR FUNCIONÁRIO ':-^80}")

    print(f"Código do Funcionário {id:>03}")

    nome = input("Nome: ").strip()
    setor = input("Setor: ").strip()

    while True:
        try:
            salario = float(input("Salário (R$): "))

            if salario <= 0:
                print("\nAtenção! Digite um valor maior que zero.\n")
            else:
                break

        except ValueError:
            print("\nAtenção! Digite um número válido.\n")

    funcionarios[id] = [nome, setor, salario]


def remover_funcionario():
    """
    Solicita que o usuário insira o ID e, se esse número de ID estiver no dicionário, remove-o
    do dicionário
    """

    print("*" * 80)
    print(f"{' MENU REMOVER FUNCIONÁRIO ':-^80}")

    while True:
        try:
            codigo = int(input("\nDigite o ID do Funcionário a ser removido: "))

            if codigo not in funcionarios.keys():
                print("\nNão existe funcionário com esse código.\n")
            else:
                funcionarios.pop(codigo)

            break
        except ValueError:
            print("\nAtenção! Digite um código válido.\n")


def consultar_funcionarios():
    """
    Exibe um menu de consulta e, em seguida, chama outras funções com base na entrada do usuário
    """

    exibir_menu = True

    while True:

        try:
            if exibir_menu:
                print("*" * 80)
                print(f"{' MENU CONSULTAR FUNCIONÁRIO(S) ':-^80}")

                imprime_menu(menu_principal[2][1])

            opcao = int(input("Escolha a opção desejada: "))

            if   opcao == 1: consultar_funcionarios_todos()
            elif opcao == 2: consultar_funcionarios_por_id()
            elif opcao == 3: consultar_funcionarios_por_setor()
            elif opcao == 4: break
            else: print("\nAtenção! Você digitou uma opção inválida.\n")

            exibir_menu = True

        except ValueError:
            print("\nAtenção! Você digitou uma opção inválida.\n")
            exibir_menu = False


def consultar_funcionarios_todos():
    """
    Imprime todos os funcionarios presentes no dicionário
    """
    if len(funcionarios) > 0:
        print("\nTodos os Funcionários cadastrados:\n")

        for key, val in funcionarios.items():
            print(f"ID: {key}")
            print(f"Nome: {val[0]}")
            print(f"Setor: {val[1]}")
            print(f"Salário (R$): {val[2]}")

    else:
        print("\nNão existem Funcionários cadastrados.")

    print("")


def consultar_funcionarios_por_id():
    """
    Solicita que o usuário informe um ID, procura o ID no dicionário e imprime os valores do item 
    associado àquele ID
    """
    while True:
        try:
            codigo = int(input("\nDigite o ID do Funcionário: "))

            if codigo in funcionarios.keys():
                print(f"ID: {codigo}")
                print(f"Nome: {funcionarios[codigo][0]}")
                print(f"Setor: {funcionarios[codigo][1]}")
                print(f"Salário (R$): {funcionarios[codigo][2]}")
            else:
                print("\nNão existe Funcionário com esse ID.\n")

            print("")

            break
        except ValueError:
            print("\nDigite um ID válido.\n")


def consultar_funcionarios_por_setor():
    """
    Solicita que usuário informe o Setor, depois percorre o dicionário de funcionários e imprime os 
    funcionários associados àquele Setor
    """
    while True:
        try:
            fabricante = input("\nDigite o Setor do(s) Funcionário(s): ")

            existe_funcionario = False

            for key, val in funcionarios.items():
                if val[1].upper() == fabricante.upper():
                    print(f"ID: {key}")
                    print(f"Nome: {val[0]}")
                    print(f"Setor: {val[1]}")
                    print(f"Salário (R$): {val[2]}")

                    existe_funcionario = True

            if not existe_funcionario:
                print("\nNão existe funcionário no setor informado.\n")

            break
        except ValueError:
            print("Digite um Setor válido.")

            print("*" * 80)


funcionarios = {}

while True:
    opcao_selecionada = mostrar_menu(menu_principal)

    if opcao_selecionada == 1:            
        cod_funcionarios = len(funcionarios) + 1
        cadastrar_funcionario(cod_funcionarios)

    elif opcao_selecionada == 2:
        consultar_funcionarios()

    elif opcao_selecionada == 3:
        remover_funcionario()

    else:
        break
```
